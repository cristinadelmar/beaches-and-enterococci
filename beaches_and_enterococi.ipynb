```python
import pandas as pd
```

```python
df = pd.read_csv('recent_beach_monitoring.csv')
df
#This are the results of beach monitoring for the months of May and June 2023
```

```python
df.info()
```

```python
# I want to create a datetime column.
df["monitoring_date"] = pd.to_datetime(df["date"])
df
```

```python
# How many monitorings have been done during the last two months for each beach?
df.beach_id.value_counts()
```

```python
df.beach_id.nunique()
```

```python
# Sort by datetime
df = df.sort_values("monitoring_date", ascending=False)
df.head()
```

```python
#Filter df to get the most recent date of monitoring for each beach
df_most_recent = df.groupby("beach_id").head(1)
df_most_recent.sort_values("beach_id").head()
```

```python
df_most_recent.beach_id.nunique()
```

```python
#Filter df to get the second to most recent date of monitoring for each beach
#df_2nd_most_recent = df.sort_values('monitoring_date', ascending=False).groupby('beach_id').nth(1)
df_2nd_most_recent = df.groupby('beach_id').nth(1)
df_2nd_most_recent.sort_values('monitoring_date', ascending=False)

df_2nd_most_recent.head()
```

```python
df_2nd_most_recent.beach_id.nunique()
```

```python
#Add columns of enterococcus_colonies_per_100_mL result 
#for the previous date when a sample was taken and the date
#First sort each df by beach id
df_most_recent = df_most_recent.sort_values(by='beach_id')
df_most_recent.head()
```

```python
df_2nd_most_recent = df_2nd_most_recent.sort_values(by='beach_id')
df_2nd_most_recent.head()
```

```python
type(df_most_recent)
df_most_recent.head()
```

```python
type(df_2nd_most_recent)
df_2nd_most_recent = df_2nd_most_recent.reset_index()

df_2nd_most_recent.head()
```

```python
#Create a new df, called working df with the info I need from each df


working_df = pd.DataFrame()
working_df['beach_id'] = df_most_recent['beach_id']
working_df['beach_name'] = df_most_recent['beach_name']
working_df['latitude'] = df_most_recent['latitude']
working_df['longitude'] = df_most_recent['longitude']
working_df['last_sample_date'] = df_most_recent['monitoring_date']
working_df['last_sample_result'] = df_most_recent['enterococcus_colonies_per_100_mL']
working_df['previous_sample_date'] = df_2nd_most_recent['monitoring_date']
working_df['previous_sample_result'] = df_2nd_most_recent['enterococcus_colonies_per_100_mL']


working_df.head()

```

```python
#To check which beaches are over the Beach Action Value in the last sample result
#First, convert the values under the result columns to integers
#As an editorial decision, we'll convert the <10 values to 5, the average between 0 and 10,
#just to be able to do calculations in the columns.
#In the graphics we'll specify it is <10.
working_df["last_sample_result"] = working_df["last_sample_result"].str.replace("510", "5")
working_df.head()
```

```python
working_df["previous_sample_result"] = working_df["previous_sample_result"].str.replace("<10", "5")
working_df.head()
```

```python
working_df["last_sample_result"] = working_df["last_sample_result"].replace(2224, 224)
```

```python
#Turn the strings under the results columns into integers.
working_df["last_sample_result"] = working_df["last_sample_result"].astype(int)
working_df["previous_sample_result"] = working_df["previous_sample_result"].astype(int)
working_df.dtypes
```

```python
#Which beaches are over the Beach Action Value in the last sample result?
#Beaches that had a value of over 70.
working_df[working_df.last_sample_result > 70]
```

```python
beaches_exceeding_BAV_df = working_df[working_df.last_sample_result > 70]
beaches_exceeding_BAV_df = beaches_exceeding_BAV_df.sort_values("previous_sample_result", ascending = False)
beaches_exceeding_BAV_df.head()
```

```python
working_df[working_df.last_sample_result > 70].count()
```

```python
#Which beaches were over the Beach Action Value in the previous sample result?
#Beaches that had a value of over 70.
working_df[working_df.previous_sample_result > 70]
```
